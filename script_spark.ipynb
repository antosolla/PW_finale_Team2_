```python
from pyspark.sql.functions import *
from pyspark.sql.types import *

path = "/Volumes/workspace/default/dataset/student_burnout_dropout_dataset_2.csv"

df = spark.read.csv(path, header=True, inferSchema=True)


df_dropped = df.na.drop()   #pulizia del dataset, sono stati eliminati tutti i null

df_chained = (
    df_dropped.groupBy(col("Department"), col ("Year_of_Study"), col("Gender"), col("Burnout_Level"))
    .agg(count(col("student_id")).alias("n_student"))
    .sort(col("Department"), col("Year_of_Study"), col("Gender"), col("Burnout_Level")
))

display(df_chained)
```
