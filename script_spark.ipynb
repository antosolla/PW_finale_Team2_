```python
from pyspark.sql.functions import *
from pyspark.sql.types import *

path = "/Volumes/workspace/default/dataset/student_burnout_dropout_dataset_2.csv"

df = spark.read.csv(path, header=True, inferSchema=True)


df.na.drop()

df_chained = (
    df.filter(col("Burnout_Level") == "High")
    .groupBy(col("Department"), col ("Age"), col("Gender"))
    .agg(count(col("student_id")).alias("n_student"))
    .sort(col("n_student").desc())
)

display(df_chained)
```
