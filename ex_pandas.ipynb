```python

import pandas as pd
import numpy as np

df = pd.read_csv("student_burnout_dropout_dataset_2.csv")

df_clean = df.dropna()
df_clean.describe()
```

```python
df_clean
```

```python
df_corr = df_clean.corr(numeric_only=True)

#year of study - stress level -0.04
df_corr
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 8))

sns.heatmap(df_corr, annot=True, cmap="coolwarm", fmt=".2f")

plt.title("Matrice di correlazione")
plt.show()


# backlogs - sleep hours 0.16
# sleep hours - study hours per day -0.12
# year of study - screen time hours 0.11
# year of study- Financial_Stress_Score -0.11
# Motivation_Score - Previous_GPA  0.10
# Peer_Pressure_Score - Motivation_Score 0.12

```

```python


def stress_category(stress):
    if stress < 3:
        return "Basso"
    elif stress <=7:
        return "Medio"
    else:
        return "Alto"
    
df["Stress_Category"] = df["Stress_Level"].apply(stress_category) 
    

df_stress_year = df_clean.groupby("Year_of_Study")["Stress_Level"].mean().reset_index()

df_stress_year.columns = ["Year of Study", "Average Stress Level"]


df_stress_year["Stress_Category"] = df_stress_year["Average Stress Level"].apply(stress_category)


print("\n LIVELLO DI STRESS PER ANNO DI STUDI\n", df_stress_year)

```

```python
print(df_stress_year["Year of Study"])
print(df_stress_year["Average Stress Level"])


plt.figure(figsize=(8, 6))

# CORREZIONE: Passa l'intera colonna (senza [0]) per mostrare tutti gli anni
sns.lineplot(
    x=df_stress_year["Year of Study"],
    y=df_stress_year["Average Stress Level"],
    color="#3b7a11",  # Il verde che hai scelto
    #linewidth=2,
    marker="o",
    label="Average Stress Level",
)

# Configurazioni del grafico
plt.title("Variazione del livello di Stress negli Anni di Studio", fontsize=14)
plt.xlabel("Anno di studi", fontsize=11)
plt.ylabel("Media livello di stress", fontsize=11)

# Impostiamo i numeri dell'asse X precisi (1, 2, 3, 4) senza decimali
plt.xticks(df_stress_year["Year of Study"])

plt.legend(loc="upper left")
plt.ylim(1, 10)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


fig, axes = plt.subplots(2, 3, figsize=(18, 10))


sns.boxplot(data=df_clean, x='Year_of_Study', y='Stress_Level', ax=axes[0, 0])
axes[0, 0].set_title('Stress Level vs Year of Study')


sns.boxplot(data=df_clean, x='Dropout_Risk', y='Stress_Level', ax=axes[0, 1])
axes[0, 1].set_title('Stress Level vs Dropout Risk')


sns.boxplot(data=df_clean, x='Dropout_Risk', y='Attendance_Percent', ax=axes[0, 2])
axes[0, 2].set_title('Attendance % vs Dropout Risk')



sns.boxplot(data=df_clean, x='Burnout_Level', y='Sleep_Hours', order=['Low', 'Medium', 'High'], ax=axes[1, 0])
axes[1, 0].set_title('Sleep Hours vs Burnout Level')


sns.boxplot(data=df_clean, x='Burnout_Level', y='Previous_GPA', order=['Low', 'Medium', 'High'], ax=axes[1, 1])
axes[1, 1].set_title('GPA vs Burnout Level')

sns.boxplot(data=df_clean, x='Dropout_Risk', y='Financial_Stress_Score', ax=axes[1, 2])
axes[1, 2].set_title('Financial Stress Score vs Dropout Risk ')




plt.tight_layout()
plt.show()
```
